```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#bring the stockmarket data 
df = pd.read_csv('stocks.csv')
import seaborn as sn
from sklearn.decomposition import PCA
print(df)
```

```python
df.info()
```

```python
#Get the information of any null value in the dataset 
df.isna().sum() 
```

```python
Y = df['Ticker']
Y = df['Date']
X = df
X.drop(['Ticker'], axis = 1, inplace = True)
X.drop(['Date'], axis = 'columns', inplace = True)
X.sample(10)
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
X = std.fit_transform(X)
X
```

```python
pca_Stockmarket = PCA(n_components = 4)
PrincipalComp_Stockmarket = pca_Stockmarket.fit_transform(X)
PrincipalComp_Stockmarket
```

```python
PrincipalComp_Stockmarket = pd.DataFrame(data = PrincipalComp_Stockmarket,columns = ['AAPL', 'MSFT', 'NFLX', 'GOOG'])
PrincipalComp_Stockmarket.sample(10)
```

```python
print('PCA of Ticker is: {}'.format(pca_Stockmarket.explained_variance_ratio_))
```
