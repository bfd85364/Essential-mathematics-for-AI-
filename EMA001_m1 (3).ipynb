```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Sleep_health.csv")
df
```

```python
df=df.drop(['Gender','Occupation','BMI Category','Sleep Disorder'], axis=1)
df
```

```python
df['Quality of Sleep'] = pd.to_numeric(df['Quality of Sleep'], errors='coerce')
df['Stress Level'] = pd.to_numeric(df['Stress Level'], errors='coerce')

target = 'Stress Level'
predictors = ['Quality of Sleep']
results = []  # ê²°ê³¼ ì €ìž¥

# formual of gradient desent 
def gradient_descent(X, Y, learning_rate=0.001, epochs=1000):
    n = len(X)
    m, c = 0.1, 0.1  # ì´ˆê¸° ê°’
    for _ in range(epochs):
        Y_pred = m * X + c
        d_m = -(2/n) * sum(X * (Y - Y_pred))
        d_c = -(2/n) * sum(Y - Y_pred)
        m -= learning_rate * d_m
        c -= learning_rate * d_c
    return m, c

# model1's each variable
X = df[predictors[0]].dropna().values
Y = df[target].dropna().values
m, c = gradient_descent(X, Y)

# MSE
Y_pred = m * X + c
mse = np.mean((Y - Y_pred)**2)

results.append({
    'Variables': predictors,
    'MSE': mse,
    'Model': (m, c)
})

# linear regressor
plt.scatter(X, Y, color='orange', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel('Quality of Sleep')
plt.ylabel('Stress Level')
plt.title('Model 1: Linear Regression using Gradient Descent')
plt.legend()
plt.show()

print(f"Model 1: slope(m) = {m}, Intercept(c) = {c}, MSE = {mse}")

# additional vars
add_vars = ['Physical Activity Level', 'Blood Pressure', 'Heart Rate', 'Daily Steps']  # ìƒˆ ë³€ìˆ˜ë“¤

for var in add_vars:
    if var in df.columns:
        X = df[[predictors[0], var]].dropna().values[:, 0] 
        Y = df[target].dropna().values

        #grediant decent
        m, c = gradient_descent(X, Y)

        # pred
        Y_pred = m * X + c
        mse = np.mean((Y - Y_pred)**2)

        # save
        results.append({
            'Variables': predictors + [var],
            'MSE': mse,
            'Model': (m, c)
        })

# 3best model(accuracy)
results = sorted(results, key=lambda x: x['MSE'])
print("\nTop 3 Variable Combinations:")
for rank, res in enumerate(results[:3], start=1):
    print(f"{rank}. Variables: {res['Variables']}, MSE: {res['MSE']}")

# visualized
best_model = results[0]
X = df[best_model['Variables'][0]].dropna().values
Y = df[target].dropna().values
Y_pred = best_model['Model'][0] * X + best_model['Model'][1]

plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='green', label='Best Regression Line')
plt.xlabel(best_model['Variables'][0])
plt.ylabel('Stress Level')
plt.title('Best Model Regression')
plt.legend()
plt.show()
```
