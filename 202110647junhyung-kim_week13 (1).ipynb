```python
import numpy as np
import matplotlib.pyplot as plt

N = 30
mean = 70
sd = 10

sample = np.random.normal(loc=mean, scale=sd, size=N)

bin_width = 10
bins = np.arange(min(sample), max(sample) + bin_width, bin_width)
hist, bin_edges = np.histogram(sample, bins=bins, density=True)

plt.hist(sample, bins=bins, density=True, alpha=0.6, color='blue', edgecolor='black', label='Histogram')

x = np.linspace(mean - 4 * sd, mean + 4 * sd, 1000)
y = (1 / (sd * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / sd) ** 2)
plt.plot(x, y, color='red', label='Normal Curve')

plt.xlabel('Value')
plt.ylabel('Relative Frequency')
plt.title('Histogram and Normal Distribution Curve')
plt.legend()

plt.show()
```
